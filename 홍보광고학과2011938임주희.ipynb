**성별에 따른 감마지티피 수치 비교**

```python
#집단별 평균표 만들기
df_mpg = mpg.groupby('성별코드' , as_index = False).agg( meang = ('감마지티피', 'mean'))
df_mpg
```

```python
sns.barplot(data = df_mpg, x = '성별코드', y = 'meang')
```

```python
sns.boxplot(x='성별코드', y='감마지티피',data=mpg)
```

감마지티피란 간 기능을 나타내는 혈액검사상의 수치, 간 내의 쓸개관(담관)에 존재하는 효소로 글루타민산을 외부에 펩티드나 아미노산 등으로 옮기는 작용을 하며, 쓸개즙(담즙) 배설 장애, 간세포 장애 발생 시 혈중에 증가하게 된다.
감마지티피 수치가 높게 나오는 것은 음주로 인한 것일 수 있는데 이때, 다른 간 기능 검사 수치는 정상 수치로 나온다면 감마지티피 수치도 일정 기간 후에 정상수치로 돌아오게 된다. 그러나 다른 간 기능 검사 수치 또한 이상 수치로 나온다면 간암, 간경병 등을 의심해봐야한다.

위 그래프에 따르면 남성의 감마지티피 평균이 여성의 감마지티피 평균보다 높은 값을 보인다. 따라서 남성이 여성보다 음주를 많이 했을 것이라고 예측할 수 있다. 또한 상자그림에서 이상치를 보이는 남성의 경우 다른 간 기능 검사를 통해 간에 다른 이상이 있는 것은 아닌지 확인할 필요가 있을 것이다.


**성별에 따른 트리글리세라이드 차이**

```python
sns.boxplot(x='성별코드', y='트리글리세라이드',data=mpg)
```

트리글리세라이드란 단순지질 혹은 중성지질을 뜻하며 이는 음식으로 섭취된 과잉에너지를 저장하기 위해 만들어지는데, 평상시에는 지방세포에 저장되어 있다가 필요시에 에너지원으로 사용된다. 트리글리세라이드는 비만, 운동 부족, 흡연, 과음, 탄수화물 다량 섭취와 같은 생활습관에 의해 혹은 당뇨병, 만성신부전 등의 질환이나 스테로이드, 베타 차단제와 같은 약제에 의해서도 상승할 수 있다.

위의 그래프는 성별을 기준으로 트리글리세라이드 수치를 상자그림으로 나타낸 것이다. 이 그래프에 따르면 남성의 트리글리세라이드의 중앙값, 최대값이 여성의 트리글리세라이드 중앙값, 최대값보다 높다. 이상치 또한 남성의 이상치가 여성의 이상치보다 눈에 띄는 모습을 보인다.

이러한 결과를 보았을 때 남성이 여성보다 높은 트리글리세라이드 수치를 보이므로 남성과 여성의 생활습관 비교, 당뇨병 등의 질환 비교, 복용중인 약제 비교를 통해 틀리글리세라이드를 관리할 수 있는 방법을 찾아내어 건강관리에 도움을 줄 수 있을 것이라고 예상된다.

**식전혈당(공복혈당)이 100초과인 사람들의 이완기혈압, 수축기혈압 산점도 확인**

고혈압 환자의 상당수가 당뇨를 앓고 있다고 하여 혈당에 문제가 있는 것으로 판단되는 사람들과 혈당수치가 정상 범위에 들어오는 사람들의 이완기혈압, 수축기혈압의 상관관계를 분석하였다.

식전혈당(공복혈당) 수치가 100mg/dL을 넘어가면 혈당에 문제가 있는 것으로 판단되기 때문에 100mg/dL을 기준으로 두 집단으로 분류하여 산점도 그래프를 그렸다.

그러나 공복혈당 수치가 100mg/dl이하인 경우와 초과인 경우 모두 우상향 그래프를 포였으며 다만 공복혈당 수치가 100mg/dl 이하인 집단의 이완기혈압 최저 수치가 100mg/dl 초과인 집단에 비해 낮은 지점에서 시작된 것을 확인할 수 있었다.

이러한 결과를 보았을 때 혈당과 혈압의 관계를 파악하기 위해서는 더 많은 표본이 필요하거나, 이 외에 다른 요인들이 함께 영향을 주고받고 있는 것으로 예상된다.
